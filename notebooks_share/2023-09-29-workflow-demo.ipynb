## Workflow Demo

Just as we previously had methods that we could run, we also have workflows, these are a composition of methods (`map`,`pred`,`group`,`de`) that you choose all or a subset of, these are then chained together into a workflow, similar to a composite function. <br>
<br>
Confogiration-wise, they behave the same as the methods, meaning you can use the same kind of `yaml` files to run them. If you want to provide parameters to any of the methods that you've chained together, you specify this in the `method_params` field in your `yaml` file. You will see an example of this below.

Import necessary standard packages

```python
import tempfile
import subprocess
from pathlib import Path
```

Import evaluation module

```python
EVAL_PTH = '../src/'
import sys
if EVAL_PTH not in sys.path:
    sys.path.append(EVAL_PTH)
import eval as ev
```

Specify path to config file

```python
CONFIG_PTH = '../examples/eval_config/wf_config.yaml'
```

Read config file

we'll use a "stupid" metric that just prints the output, since this is for illustrative purposes.

```python
import yaml

with open(CONFIG_PTH, 'r') as f:
    design = yaml.safe_load(f)
    
```

Run evaluation and look at output (saved to temporary folder for demo purpose, will be removed once context closes).
<br>
<br>
**Note**: if you actually want to inspect the results, change the context from a `TemporaryDirectory` to any path of your choice.

```python
with tempfile.TemporaryDirectory() as tmpdir:
    ev.run.run(design,tmpdir)

    subprocess.run(['tree',tmpdir])
    res_files = list(Path(tmpdir).rglob('*.txt'))
    for fn in res_files:
        print('\n Content of: {}'.format(fn))
        subprocess.run(['cat',fn])
```
